```python
def get_files():
    files = os.listdir()
    for file in files:
        if not file.endswith('.csv'):
            files.remove(file)
    return files
```

```python
def get_student_details(data):
    attendence = data['attendance_records'].copy()
    student_scores = data['student_scores'].copy()
    students = data['Students']
    final = students.merge(attendence , on = "student_id").merge(student_scores, on = "student_id")
    attendence.replace(to_replace = "P", value = 1, inplace = True)
    attendence.replace(to_replace = "A", value = 0, inplace = True)
    attendence.index = attendence["student_id"]
    attendence.drop(columns = "student_id", inplace = True)
    student_scores.index = student_scores["student_id"]
    student_scores.drop(columns = "student_id", inplace = True)
    avg_attendence = attendence.mean(axis = 1).to_frame()
    avg_attendence.rename(columns = {0: "attendance"}, inplace = True)
    final = final.merge(avg_attendence, on = "student_id")
    s_s = student_scores.mean(axis = 1).to_frame()
    s_s.rename(columns = {0: "avg_score"}, inplace = True)
    final = final.merge(s_s, on = "student_id"  )
    return final.to_json()
```
